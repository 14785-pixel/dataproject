```python
import numpy as np
import pandas as pd
import pymannkendall as mk
import matplotlib.pyplot as plt
import statsmodels.api as sm

rainfall_data = pd.read_csv("All_India_rainfall_act_dep_1901_2015.csv",header = 0,index_col = 0)
```

```python
rainfall_data.head()
```

```python
rainfall_data.plot()
```

```python

fig, ax = plt.subplots(figsize=(12, 8))
sm.graphics.tsa.plot_acf(rainfall_data, lags=10, ax=ax);
```

```python
mk.original_test(rainfall_data,alpha = 0.05)
```
